```python

import numpy as np
import matplotlib.pyplot as plt
x,y1,y2,y3,y4 = np.loadtxt('melting.txt',skiprows=1,usecols=(0,1,2,3,4) ,unpack=True)
y5=np.loadtxt('melting.txt',skiprows=1,usecols=(5) ,unpack=True)

def new_meaning(y):
   min=np.min(y)
   max=np.max(y)
   for i in range(len(y)):
    if y[i]==min:
     y[i]=0
    if y[i]==max:
        y[i]=1
    else:
       y[i]= (y[i]-min)/(max-min)
       if y[i]<=0:
           y[i]=0
def graph(y):
 plt.plot(x,y)
 plt.xlabel('time')
 plt.ylabel('signal')

cords=[y1,y2,y3,y4,y5]
for y in cords:
    y=new_meaning(y)
    
for y in cords:
    graph(y)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
a=np.loadtxt('ex4.txt', unpack=True)
average=np.mean(a)
std_dev=np.std(a)
print('среднее:',average,'стандартное отклонение:', std_dev)

a.sort()
a_new=[]
x = [i for i in np.quantile(a, [0.25, 0.5, 0.75])]

for i in a:
    if i > (x[0] - 1.5 * (x[2] - x[0])) and i < (x[2] + 1.5 * (x[2] - x[0])):
        a_new.append(i)
average_2=np.mean(a_new)
std_dev_2=np.std(a_new)
print('среднее без выбросов:',average_2,'стандартное отклонение без выбросов:', std_dev_2)
```
